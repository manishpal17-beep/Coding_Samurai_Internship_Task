```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($)')
plt.scatter(df.area,df.price, color='green' , marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[2600]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
m=135.78767123
x=2600
b=180616.43835616432

y = m*x + b
135.78767123*2600+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='green' , marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='yellow')
```
